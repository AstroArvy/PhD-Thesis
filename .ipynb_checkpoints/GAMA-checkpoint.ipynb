```python
import astropy as ap
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad as integral
```

```python
from astropy.table import Table

# Load the fits file
data = Table.read('StellarMassesGKVv24.fits', format='fits')

filter_data = data[(data['logmstar'] > 9) & (data['Z'] > 5e-3) & (data['Z'] < 0.065)]
```

```python
z = filter_data['Z']
logM = filter_data['logmstar']
```

```python
counts, massbin = np.histogram(logM, bins=100)
```

# Schechter Mass Function

\begin{equation}
    \Phi (log M) d (log M) = ln(10) \cdot \phi^{*} 10^{log\frac{M}{M^{*}}(\alpha + 1)} \times exp(-10^{log\frac{M}{M^{*}}}) d (log M)
\end{equation}

```python
from scipy.optimize import curve_fit

counts = counts
massbins= massbin #log mass

# Calculate mass midpoints
mass_midpoints = 0.5 * (massbins[:-1] + massbins[1:]) #log mass

# # Schechter function definition
# def schechter_func(mass, phi_star, m_star, alpha):
#     x = mass / m_star
#     return phi_star * (x**alpha) * np.exp(-x)

# Schechter function definition
def schechter_func(mass, phi_star, m_star, alpha):
    x = np.log(mass / m_star)
    return np.log(10)*phi_star*(10**(x*(alpha+1)))*np.exp(-10**x)


# Fitting the Schechter function to the data
params, cov = curve_fit(schechter_func, 10**mass_midpoints, counts, p0=[1e-3, 1e11, -1.3], maxfev=10000)

# Unpacking the fitted parameters
phi_star_fitted, m_star_fitted, alpha_fitted = params

# Generating masses for plotting the fitted function
masses_for_plot = np.linspace(min(mass_midpoints), max(mass_midpoints), 1000)

# Evaluating the fitted Schechter function over the generated masses
fitted_counts = schechter_func(10**masses_for_plot, *params)

# Plotting the histogram of galaxy counts and the fitted Schechter function
plt.figure(figsize=(10, 7))
plt.bar(mass_midpoints, counts, width=np.diff(massbins), color='gray', alpha=0.5, label='Galaxy Counts')
plt.plot(masses_for_plot, fitted_counts, color='red', linewidth=2, label='Fitted Schechter Function')
#plt.xscale('log')
plt.yscale('log')
plt.ylim(1,5000)
plt.xlabel('Galaxy Mass (M)')
plt.ylabel('Number of Galaxies')
plt.title('Galaxy Mass Distribution and Fitted Schechter Function')
plt.legend()
plt.show()

# Displaying the fitted parameters
phi_star_fitted, m_star_fitted, alpha_fitted
```

# Likelihood Function

\begin{equation}
L = \frac{\prod_{m_i} \Phi_{sch}(M_i|\phi^*,M^*,\alpha)}{{\int_{M_{min}}^{M_{max}}}\Phi d(logM)}
\end{equation}

\begin{equation}
\log L = \sum_{m_i} \log \Phi_{sch}(M_i|\phi^*,M^*,\alpha) - N (\log \int_{M_{min}}^{M_{max}}\Phi d(logM))
\end{equation}

```python
# Schechter function definition
def schechter_func(mass, phi_star, m_star, alpha):
    x = np.log10(mass / m_star)
    return np.log(10)*phi_star*(10**(x*(alpha+1)))*np.exp(-10**x)

def Likelihood(mass, phi_star, m_star, alpha):
    sch = schechter_func(10**mass, phi_star, m_star, alpha) #10^mass because mass we use is logM
    
    #initiate the integral function
    func = lambda x: np.log(10)*phi_star*(10**((x-np.log10(m_star))*(alpha+1)))*np.exp(-10**(x-np.log10(m_star)))
    norm = integral(func, a=min(logM), b=max(logM)) #integral = scipy.integrate.quad
    
    #define a very small constant to avoid log(0) in calculations
    const = 1e-10
    numerator=0
    for i in range(len(mass)):
        numerator += np.log(sch[i]+const)
    
    denominator = (len(mass)*np.log(norm[0]+const))

    L = numerator - denominator
    
    return L
```

```python
phi_star=np.linspace(0,50, 10)

M_star=np.logspace(8,13,10)

alpha = np.linspace(-1.5,-1,100)
```

```python
Likelihood(logM, 32.522501276810594, 86751986104.16612, -1.1227156801572107 )
```

```python
L_values = np.zeros(len(alpha))

from tqdm import tqdm
for i in tqdm(range(len(alpha))):
    phi, M = 32.522501276810594, 86751986104.16612
    L_values[i] = Likelihood(logM, phi, M, alpha[i])
```

```python
plt.figure(figsize=(7,6.2))

minima = alpha[np.argmax(L_values)]

plt.plot(alpha,L_values, color= 'blue')

plt.axvline(x = minima, color = 'black',linestyle=':', label = 'axvline - full height')

plt.text(minima-0.005, plt.gca().get_ylim()[0]-7, str(round(minima,2)), ha='center', va='top', fontweight=1000)

plt.title(r'$\alpha$ vs $\log L$')

plt.grid(alpha=0.5, linestyle=':')

plt.xlabel('alpha', fontsize=13)
plt.ylabel('log L', fontsize=13)

plt.show()
```
